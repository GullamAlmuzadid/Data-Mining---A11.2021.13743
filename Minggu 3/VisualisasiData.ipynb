```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("movies.csv")
data.head()
```

```python
data.info()
```

```python
data['Certificate'].unique()
```

```python
rating_certificate = 'R'
data[data['Certificate']==rating_certificate]
```

```python
unique_countries = data[data.Year>2020]['Filming_location'].unique()
unique_countries.sort()
unique_countries
```

```python
country_year_release = []
for i in unique_countries:
  country_year_release.append(data[data.Year>2020][data['Filming_location']==i][data['Certificate']==rating_certificate].Year.sum())
```

```python
print(set(zip(unique_countries, country_year_release)))
```

```python
unique_month = data['Month'][data.Year>2021].unique()
unique_month
```

```python
month_year_release = []
for i in unique_month:
  month_year_release.append(data[data.Year>2020][data['Month']==i][data['Certificate']==rating_certificate].Year.sum())
```

```python
print(set(zip(unique_month, month_year_release)))
```

```python
nan_indices = []
for i in range(len(unique_month)):
  if type(unique_month[i])==float:
    nan_indices.append(i)
```

```python
unique_month = list(unique_month)
month_year_release = list(month_year_release)
```

```python
for i in nan_indices:
  unique_month.pop(i)
  month_year_release.pop(i)

unique_month
```

```python
plt.barh(unique_countries, country_year_release)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_countries, country_year_release)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(country_year_release)
plt.legend(unique_countries, loc='best')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.pie(month_year_release)
plt.legend(unique_month, loc='best')
plt.show()
```
